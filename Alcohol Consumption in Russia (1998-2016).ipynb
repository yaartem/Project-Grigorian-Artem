```python
import warnings
warnings.simplefilter('ignore')
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
train_df = pd.read_csv("russia_alcohol.csv - Лист1.csv")
train_df.head(20)
```

```python
print('Mean value of vodka consumption in one year:', train_df['vodka'].mean(), ("litres per capita"))
print('Median of vodka consumption in one year:', train_df['vodka'].median(), ("litres per capita"))
print('Standart deviation of vodka consumption in one year:', train_df['vodka'].std(), ("litres per capita"))
```

```python
print('Mean value of beer consumption in one year:', train_df['beer'].mean(), ("litres per capita"))
print('Median of beer consumption in one year:', train_df['beer'].median(), ("litres per capita"))
print('Standart deviation of beer consumption in one year:', train_df['beer'].std(), ("litres per capita"))
```

```python
print('Mean value of brandy consumption in one year:', train_df['brandy'].mean(), ("litres per capita"))
print('Median of brandy consumption in one year:', train_df['brandy'].median(), ("litres per capita"))
print('Standart deviation of brandy consumption in one year:', train_df['brandy'].std(), ("litres per capita"))
```

```python
plt.figure(figsize=(16,20))
plt.scatter(train_df['vodka'], train_df['region'])
plt.xlabel('Litres per capita')
plt.ylabel('Regions')
plt.title('Consumption of vodka in different regions')
plt.show()
```

```python
plt.figure(figsize=(16,30))
alotvodka=train_df.query("vodka >= 11.818693693693698")
plt.scatter(alotvodka["vodka"], alotvodka["region"])
plt.xlabel('Vodka')
plt.ylabel('Regions')
plt.title('Consumption of vodka more than mean')
plt.xticks()
plt.show()
```

```python
plt.figure(figsize=(16,30))
alotbeer=train_df.query("beer >= 51.260147719974235")
plt.scatter(alotbeer["beer"], alotbeer["region"])
plt.xlabel('Beer')
plt.ylabel('Regions')
plt.title('Consumption of beer more than mean')
plt.xticks()
plt.show()
```

```python
plt.figure(figsize=(16,30))
alotbrandy=train_df.query("brandy >= 0.5269980632666229")
plt.scatter(alotbrandy["brandy"], alotbrandy["region"])
plt.xlabel('Brandy')
plt.ylabel('Regions')
plt.title('Consumption of brandy more than mean')
plt.xticks()
plt.show()
```

```python
print("TOP10 Most vodka drinking regions")
alotvodka=train_df.query("vodka >= 11.818693693693698")
print(alotvodka["region"].value_counts()[0:10])
print()
print("TOP10 Most beer drinking regions")
alotbeer=train_df.query("beer >= 51.260147719974235")
print(alotbeer["region"].value_counts()[0:10])
print()
print("TOP10 Most brandy drinking regions")
alotbrandy=train_df.query("brandy >= 0.5269980632666229")
print(alotbrandy["region"].value_counts()[0:10])
print()
```

```python
Moscow = train_df.query("region=='Moscow' and year==2016")
print(Moscow)
Moscow['overall'] = Moscow['wine'] + Moscow['beer'] + Moscow['vodka'] + Moscow['champagne'] + Moscow['brandy']
wineshare = float(Moscow['wine']/Moscow['overall'])*100
beershare = float(Moscow['beer']/Moscow['overall'])*100
vodkashare = float(Moscow['vodka']/Moscow['overall'])*100
champshare = float(Moscow['champagne']/Moscow['overall'])*100
brandshare = float(Moscow['brandy']/Moscow['overall'])*100
lablz='Wine', 'Beer', 'Vodka', 'Champagne', 'Brandy'
plt.pie([ wineshare,beershare, vodkashare, champshare, brandshare], labels=lablz)
```
